```python
# Packages
import json
import pandas as pd

# Load JSON data from file
with open('../../Steam-Data/data/raw/steam_game_data.json', 'r') as file:
    game_data = json.load(file)
```

```python
# Extract publisher game data
publisher_game_data = [	{	'app_id'			: app_id,
							'name'				: game['name'],
							"publisher"			: publisher, 
							"genre"				: genre,
							"positive_reviews"	: game["positive"],
							"negative_reviews"	: game["negative"],
							"metacritic_score"	: game["metacritic_score"],
							"release_date"		: game["release_date"],
							"price"				: game["price"],
							"estimated_owners"	: game["estimated_owners"],
							"peak_ccu"			: game["peak_ccu"]
						}
					for app_id, game in game_data.items() for publisher in game['publishers'] for genre in game['genres']
				]


# Extract developer game data
developer_game_data = [{	'app_id'			: app_id,
							'name'				: game['name'],
							"developer"			: developer, 
							"genre"				: genre,
							"positive_reviews"	: game["positive"],
							"negative_reviews"	: game["negative"],
							"metacritic_score"	: game["metacritic_score"],
							"release_date"		: game["release_date"],
							"price"				: game["price"],
							"estimated_owners"	: game["estimated_owners"],
							"peak_ccu"			: game["peak_ccu"]
						}
	                    for app_id, game in game_data.items() for developer in game['developers'] for genre in game['genres']
					]
```

```python
# Save data to CSV
pd.DataFrame(publisher_game_data).to_csv('../data/publisher_game_data.csv', index=False)
pd.DataFrame(developer_game_data).to_csv('../data/developer_game_data.csv', index=False)
```
